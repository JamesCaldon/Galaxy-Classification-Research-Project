```python
"""

This is for morphological classification of galaxies by CNN,
By Kenji Bekki, on 2017/11/15
Revised on 2020/2/14 (Nair & Abraham 2010)
For test only.

"""

from keras.models import model_from_json
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
mpl.rcParams['figure.figsize'] = [10, 10]
mpl.rcParams['figure.dpi'] = 72
```

```python
%run ../utility/data_loading.ipynb
#x, Y, metadata = load_data(name="NGC_6861")

#plt.imshow(x)
from PIL import Image
from PIL import ImageOps
img = Image.open(r"..\data\fits_testing\NGC6861.jfif")
#img.show()
img = ImageOps.grayscale(img)
img = ImageOps.invert(img)
img = img.resize((100, 100))
x = np.array(img)
x = (x - np.min(x)) / (np.max(x) - np.min(x))
print(np.max(x))
x = x.reshape([-1, 100, 100, 1])
#plt.imshow(x[0])
```

```python
# load json and create model
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model.h5")
print("Loaded model from disk")
y_pred_prob = loaded_model.predict(x)
y_pred = y_pred_prob.argmax(axis=1)
```
